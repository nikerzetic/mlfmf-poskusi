```python
def test_embeddings_coverage():
    import helpers

    rows = 0
    missing_in_label2raw = []
    missing_in_embeddings = []
    included_in_embeddings = {}
    inclussion_incidence = {}
    number_of_unks = 0
    _, label2raw = helpers.create_dictionaries("stdlib")
    for key, _ in label2raw.items():
        label2raw[key] = 0
    print("Analizing data...")
    with open("data/embeddings/code2class/stdlib.tsv", "r", encoding="utf-8") as f:
        f.readline()
        for line in f:
            rows += 1
            name = line.strip("\n").split("\t")[0]
            if not name in label2raw:
                if name == "<unk>":
                    number_of_unks += 1
                    continue
                missing_in_label2raw.append(name)
                continue
            label2raw[name] += 1
            included_in_embeddings[name] = included_in_embeddings.get(name, 0) + 1
    for key, value in label2raw.items():
        if value == 0:
            missing_in_embeddings.append(key)
    print("Total rows in stdlib.tsv:", rows)
    print("Total entries in label2raw: ", len(label2raw))
    print("In stdlib.tsv, but not in label2raw:", len(missing_in_label2raw))
    print("In label2raw but not in stdlib.tsv: ", len(missing_in_embeddings))
    print("Number of <unk>s:", number_of_unks)
    # print("Included in stdlib.tsv:")
    # for key, value in included_in_embeddings.items():
    #     inclussion_incidence[value] = inclussion_incidence.get(value, 0) + 1
    #     print(f"\t{key}: {value}")
    # print("Number of labels that repeat x times:")
    # for key, value in inclussion_incidence.items():
    #     print(f"\t{key}: {value}")
    return missing_in_embeddings, missing_in_label2raw
```
