```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def syntax_tree_from_dag(dag_file_path):
    G = nx.DiGraph()
    edges = {}
    with open(dag_file_path, "r", encoding="utf-8") as file:
        file.readline()
        for line in file:
            parts = line.split("\t")
            node_id = int(parts[0])
            node_type = parts[1].replace(":", "")
            node_description = parts[2]
            node_children = eval(parts[3])

            G.add_node(
                node_id, 
                type=node_type, 
                desc=node_description
                )
            edges[node_id] = node_children
    for u, children in edges.items():
        for v in children:
            G.add_edge(u,v)
    return G

def draw_from_dag(dag_file_path):
    G = syntax_tree_from_dag(dag_file_path)
    type_labels = nx.get_node_attributes(G, "type")
    desc_labels = nx.get_node_attributes(G, "desc")

    # pos = nx.planar_layout(G)
    pos = nx.nx_pydot.graphviz_layout(G, prog="dot")
    # pos = nx.kamada_kawai_layout(G)
    # pos = nx.spring_layout(G, iterations=50, seed=42)
    state_pos = {n: (x, y-8) for n, (x,y) in pos.items()}

    plt.figure(figsize=(20, 20), dpi=100)
    nx.draw_networkx(G, pos, labels=type_labels, node_size=1000, font_size=9)
    nx.draw_networkx_labels(G, state_pos, labels=desc_labels, font_color='red')
```
